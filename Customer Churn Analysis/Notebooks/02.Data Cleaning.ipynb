```python
import pandas as pd

df = pd.read_excel("Telco_customer_churn.xlsx")
df.head()
```

```python
df = df.drop(columns=[
    'CustomerID',
    'Count',
    'Country',
    'State',
    'City',
    'Zip Code',
    'Lat Long',
    'Latitude',
    'Longitude'
])
```

```python
df.shape
```

```python
df['Total Charges']
```

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
df['Total Charges'].isnull().sum()
```

```python
df['Total Charges'].isnull().sum()
```

```python
df = df[df['Total Charges'].notnull()]
df.shape
```

Handling Missing Billing Values

After converting Total Charges to numeric, 11 records contained missing values.
These correspond to customers with no billing history (new customers).
Only these rows were removed to preserve the rest of the dataset and avoid biasing the churn analysis.

```python
df['Multiple Lines'].unique()
```

```python
df['Multiple Lines'].value_counts(dropna=False)
```

```python
yes_no_cols = [
    'Partner','Dependents','Phone Service','Paperless Billing'
]

for col in yes_no_cols:
    df[col] = df[col].map({'Yes':1,'No':0})
```

```python
df['Partner'].value_counts()
```

```python
df = df.drop(columns=['Churn Label','Churn Score','Churn Reason'])
```

#The churn score was removed to prevent data leakage because it represents an existing prediction generated by the company rather than customer behaviour.

```python
df.columns
```

```python
df.to_csv("churn_clean.csv", index=False)
```

Data Cleaning Summary

Some irrelevant features and location columns were removed, as they do not have any direct influence on customer behaviour.
The Total Charges column was converted to numeric, and invalid records were removed.
Binary categorical columns were encoded into numeric values, and leakage variables such as churn score and churn reason were removed. This was done to eliminate data leakage and bias.
The final dataset now contains only behavioural and billing features suitable for churn analysis and modelling.
